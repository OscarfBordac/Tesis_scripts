```python
#Librerias

import rasterio
import pandas as pd
import numpy as np
import os
```

```python
carpeta =r"D:\Trabajo\Tesis\Carpeta indices"
```

```python
def calculo_ndvi(imagen, nodata=0):
  with rasterio.open(imagen) as src:
    # Read bands (assuming Landsat-like band order)
    nir = src.read(5).astype(float)
    red = src.read(4).astype(float)

    # Check for nodata values and replace with the specified value
    mask = nir != src.nodata  # Create a mask based on nodata value
    nir = np.where(mask, nir, nodata)
    red = np.where(mask, red, nodata)

    # Calculate NDVI, handling potential division by zero with np.nan_to_num
    ndvi = np.nan_to_num((nir - red) / (nir + red))  # Prevent division by zero errors

    # Update profile with desired parameters
    profile = src.profile
    profile.update(
        dtype=rasterio.float32,
        count=1,
        compress='lzw'
    )

    # Generate output filename
    nombre_salida = os.path.splitext(os.path.basename(imagen))[0]
    salida = f'D:/Trabajo/Tesis/Carpeta indices/Indices/NDVI/{nombre_salida}_ndvi2.tif'

    # Write NDVI to new file
    with rasterio.open(salida, 'w', **profile) as dst:
      dst.write(ndvi, 1)
def calculo_savi(imagen, L=0.5):
  with rasterio.open(imagen) as src:
    # Leer bandas NIR y rojo
    nir = src.read(5).astype(float)
    red = src.read(4).astype(float)

    # Crear máscara para nodata (ajustar según el valor nodata de tu imagen)
    nodata_mask = nir == src.nodata
    nir = np.where(nodata_mask, 0, nir)  # Reemplazar nodata con 0
    red = np.where(nodata_mask, 0, red)

    # Calcular SAVI
    savi = ((nir - red) / (nir + red + L)) * (1 + L)

    # Actualizar perfil para la salida
    profile = src.profile
    profile.update(
        dtype=rasterio.float32,
        count=1,
        compress='lzw'
    )

    # Generar nombre de salida
    nombre_salida = os.path.splitext(os.path.basename(imagen))[0]
    salida = f'D:/Trabajo/Tesis/Carpeta indices/Indices/SAVI/{nombre_salida}_savi.tif'

    # Escribir SAVI a un nuevo archivo
    with rasterio.open(salida, 'w', **profile) as dst:
      dst.write(savi, 1)
def calculo_nbr(imagen):
  with rasterio.open(imagen) as src:
    # Leer bandas NIR y SWIR
    nir = src.read(5).astype(float)
    swir = src.read(7).astype(float)

    # Crear máscara para nodata (ajustar según el valor nodata de tu imagen)
    nodata_mask = nir == src.nodata
    nir = np.where(nodata_mask, 0, nir)  # Reemplazar nodata con 0
    swir = np.where(nodata_mask, 0, swir)

    # Calcular NBR
    nbr = (nir - swir) / (nir + swir)

    # Actualizar perfil para la salida
    profile = src.profile
    profile.update(
        dtype=rasterio.float32,
        count=1,
        compress='lzw'
    )

    # Generar nombre de salida
    nombre_salida = os.path.splitext(os.path.basename(imagen))[0]
    salida = f'D:/Trabajo/Tesis/Carpeta indices/Indices/NBR/{nombre_salida}_nbr.tif'

    # Escribir NBR a un nuevo archivo
    with rasterio.open(salida, 'w', **profile) as dst:
      dst.write(nbr, 1)
def calculo_nbr2(imagen):
  """
  Calcula el Índice de Quemadura Normalizado 2 (NBR2) para una imagen raster.

  Args:
      imagen (str): Ruta al archivo de imagen raster.

  Returns:
      None
  """

  with rasterio.open(imagen) as src:
    # Leer bandas SWIR1 y SWIR2
    swir1 = src.read(6).astype(float)  # Asumiendo SWIR1 en la banda 6
    swir2 = src.read(7).astype(float)  # Asumiendo SWIR2 en la banda 7

    # Crear máscara para nodata (ajustar según el valor nodata de tu imagen)
    nodata_mask = swir1 == src.nodata
    swir1 = np.where(nodata_mask, 0, swir1)
    swir2 = np.where(nodata_mask, 0, swir2)

    # Calcular NBR2
    nbr2 = (swir1 - swir2) / (swir1 + swir2)

    # Actualizar perfil para la salida
    profile = src.profile
    profile.update(
        dtype=rasterio.float32,
        count=1,
        compress='lzw'
    )

    # Generar nombre de salida
    nombre_salida = os.path.splitext(os.path.basename(imagen))[0]
    salida = f'D:/Trabajo/Tesis/Carpeta indices/Indices/NBR2/{nombre_salida}_nbr2.tif'

    # Escribir NBR2 a un nuevo archivo
    with rasterio.open(salida, 'w', **profile) as dst:
      dst.write(nbr2, 1)

def calculo_ndwi(imagen, nodata=0):
  with rasterio.open(imagen) as src:
    # Leer bandas NIR y verde (ajusta los índices según tu imagen)
    nir = src.read(5).astype(float)  # Asumiendo que NIR está en la banda 5
    green = src.read(3).astype(float)  # Asumiendo que verde está en la banda 3

    # Crear máscara para nodata
    mask = nir != src.nodata
    nir = np.where(mask, nir, nodata)
    green = np.where(mask, green, nodata)

    # Calcular NDWI
    ndwi = np.where((nir + green) == 0, 0, np.nan_to_num((nir - green) / (nir + green)))

    # Crear perfil para la salida
    profile = src.profile
    profile.update(
        dtype=rasterio.float32,
        count=1,
        compress='lzw'
    )

    # Generar nombre de salida
    nombre_salida = os.path.splitext(os.path.basename(imagen))[0]
    salida = f'D:/Trabajo/Tesis/Carpeta indices/Indices/NDWI/{nombre_salida}_ndwi.tif'

    # Escribir NDWI a un nuevo archivo
    with rasterio.open(salida, 'w', **profile) as dst:
      dst.write(ndwi, 1)

lista_tif = []

for imagen in os.listdir(carpeta):
    if imagen.endswith('.tif'):
        ruta_imagen=os.path.join(carpeta,imagen)
        calculo_ndvi(ruta_imagen)
        calculo_savi(ruta_imagen)
        calculo_nbr(ruta_imagen)
        calculo_nbr2(ruta_imagen)
        calculo_ndwi(imagen, nodata=0)
        lista_tif.append(ruta_imagen)
        print(ruta_imagen)
```
